```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
```

```python
"""
Read in data
"""

# sold data
sold = pd.read_json('sold2.jl',lines=True)

# sale data
sale = pd.read_json('sale2.jl',lines=True)
```

```python
"""
quick clean of data
"""
# list of filtering parameters
clean = ['url','address','city_state','price','address','bedrooms','bathrooms','area']#,'year_built','lot_size']

# drop duplicate rows
sale = sale.drop_duplicates(clean)

# drop data with missing data in clean
sale = sale.dropna(subset = clean)

#resulting size
print(str(sale.shape[0]) + ' viable house sale data points')
```

```python
"""
Check initial correlations
"""

# correlation parameters
clean = ['price','address','bedrooms','bathrooms','area']

# correlation matrix

corr = sale[clean].corr()

#initialize figure
fig, ax = plt.subplots(1,1, figsize = (5, 5), dpi=300)

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns,
       cmap="cividis")
```
